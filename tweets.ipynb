```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load the DataFrame from the provided URL
df = pd.read_csv('https://github.com/KAILASHVenkat/Week2TextToneTune/raw/main/sentiment_tweets3.csv')

# Select the 'message to examine' column and rename it to 'tweets'
df = df[['message to examine']]
df.rename(columns={'message to examine': 'tweets'}, inplace=True)

# Initialize Lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Tokenization, stop words removal, and lemmatization function
def preprocess_tweet(tweet):
    # Tokenization
    tokens = word_tokenize(tweet)

    # Remove stopwords and perform lemmatization
    processed_tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens if token.isalnum() and token.lower() not in stop_words]

    # Combine tokens back into a sentence
    processed_tweet = ' '.join(processed_tokens)

    return processed_tweet

# Apply preprocessing to the 'tweets' column and store the results in a new column 'processed_tweets'
df['processed_tweets'] = df['tweets'].apply(preprocess_tweet)

# Vectorization using TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['processed_tweets'])

# Create a new DataFrame (df1) with TF-IDF features
df1 = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Display the head of the new DataFrame df1
#df1.head()
df
```
