```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sn
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KAILASHVenkat/Week1TextToneTune/main/drug200.csv')
```

```python
Sex_Mapping={"M":0,"F":1}
df["Sex"]=df["Sex"].map(Sex_Mapping)
bp_map={"HIGH":2,"NORMAL":1,"LOW":0}
df["BP"]=df["BP"].map(bp_map)
c_map={"HIGH":1,"NORMAL":0}
df["Cholesterol"]=df["Cholesterol"].map(c_map)
d_map={"DrugY":0,"drugA":1,"drugB":2,"drugC":3,"drugX":4}
df["Drug"]=df["Drug"].map(d_map)
df
```

```python
x=df[['Age','Sex','BP','Cholesterol','Na_to_K']]
y=df['Drug']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model1 = keras.Sequential([
    keras.layers.Dense(64, input_shape=(5,), activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(5, activation='softmax')
])

optimizer = keras.optimizers.Adam(learning_rate=0.01)  # Adjust the learning rate as needed
model1.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model1.fit(x_train_scaled, y_train, epochs=20, validation_data=(x_test_scaled, y_test))
```
